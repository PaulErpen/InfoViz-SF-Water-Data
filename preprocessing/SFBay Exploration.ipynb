```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
df_sfbay = pd.read_csv("SFBay.csv", sep=";")
```

```python
display(df_sfbay)
```

```python
display(df_sfbay[df_sfbay["Station.Number"] == 4.0].copy(deep=True))
```

```python
df_sfbay["MonthYear"] = df_sfbay["TimeStamp"].apply(lambda x: x[:7])
df_sf_bay_grouped_4 = df_sfbay[(df_sfbay["Station.Number"] == 4.0) 
                               & (df_sfbay["Depth"] == 2.0)].copy(deep=True).groupby(["MonthYear"]).mean()
display(df_sf_bay_grouped_4)
```

```python
for column in df_sf_bay_grouped_4.columns:
    if column == "TimeStamp" or column == "Depth" or column == "Distance.from.36" or column == "Station.Number":
        continue
    else:
        plt.plot(df_sf_bay_grouped_4.index, df_sf_bay_grouped_4[column])
        ax = plt.gca()
        ax.xaxis.set_major_locator(ticker.MaxNLocator(22))
        plt.title(column)
        plt.xticks(rotation=90)
        plt.savefig("figures/{}.png".format(column), dpi=300)
        plt.show()
```

```python
station_numbers = np.sort([int(i) for i in df_sfbay["Station.Number"].unique().tolist()])
print(station_numbers)
```

```python
for station_number in station_numbers:
    df_for_station = df_sfbay[df_sfbay["Station.Number"] == station_number].copy(deep=True).groupby(["MonthYear"]).mean()
    display(df_for_station)
    break
```
